# Importing Data

## Loading and Displaying DataFrame



Importing the necessary libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import display, HTML
import warnings
warnings.filterwarnings("ignore")
```

Reading the CSV file into a dataFrame and displaying the first few rows of the DataFrame to examine its structure and contents

```python
df = pd.read_csv('imputed_dataset.csv')
pd.set_option('display.max_columns', None)
df.head()
```

Checking the dimensions of the dataframe

```python
df.shape
```

Re-evaluating for missing values

```python
column_missing_info = []
for col in df.columns:
  if(df[col].isnull().any()):
    column_missing_info.append({
        'Column': col,
        'Data Type': df[col].dtype,
        'Null values count': df[col].isnull().sum(),
        '% of missing data': ((df[col].isnull().sum())/len(df))*100
    })

if column_missing_info:
  column_info_df = pd.DataFrame(column_missing_info)
  display(HTML(column_info_df.sort_values(by=["Data Type", "Null values count"]).to_html()))
else:
  print('No missing values!')
```

# Data Transformation

Dropping 'hospital_id', 'icu_id' columns as they didn't have any considerable correlation with the target variable

```python
df = df.drop(columns=['hospital_id', 'icu_id'])
df.head()
```

Separating and Storing numerical and categorical columns

```python
numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns
categorical_cols = df.select_dtypes(include=['object']).columns
```

Taking copy of orginal dataframe before making any transformations

```python
df_original = df.copy(deep=True)
```

## Categorical encoding

Performing one-hot encoding on categorical variables and assigning the new encoded columns to the dataframe

```python
discret_df = pd.get_dummies(df[categorical_cols], drop_first = True)
df = pd.concat([df, discret_df], axis=1)
df.drop(categorical_cols, axis=1, inplace=True)
df.replace({True: 1, False: 0}, inplace=True)

display(HTML(df.head().to_html()))
```

Separating numeric columns from binary columns, so that standardization can be applied to numeric columns only

```python
binary_columns = [x for x in df.columns if (df[x].nunique() == 2 and set(df[x].unique()) == {0, 1})]
numeric_cols = [x for x in df.columns if x not in binary_columns]
```

## Performing Data Normalization

Performing Data Normalization to ensure that all numerical features in the dataset have a similar scale

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
df[numeric_cols] = pd.DataFrame(scaler.fit_transform(df[numeric_cols]), columns=numeric_cols)
display(HTML(df.head().to_html()))
```

# Split Dataset

Separating the target feature from the independent features for model training

```python

df_features = df.drop('diabetes_mellitus', axis=1)
display(df_features.head(5))
```

## K Means Clustering

The goal of clustering is to group the dataset into diabetes_mellitus positive and negative clusters. Hence, n_clusters value has been predetermined as 2. (Elbow method to vizualize and identify value of k is not required)

```python
from sklearn.cluster import KMeans

kmeans=KMeans(n_clusters=2, random_state=10,max_iter=20000)
kmeans.fit(df_features)
y_pred=kmeans.predict(df_features)

# Getting the cluster labels and add it back to the original data
opt=kmeans.labels_
df_original['Pred_Cluster']=opt
display(df_original.head(5))
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score

actual_values = df_original['diabetes_mellitus']
predicted_values = df_original['Pred_Cluster']  

conf_matrix = confusion_matrix(actual_values, predicted_values)
accuracy = accuracy_score(actual_values, predicted_values)
precision = precision_score(actual_values, predicted_values)
recall = recall_score(actual_values, predicted_values)
f1 = f1_score(actual_values, predicted_values)


print("Confusion Matrix:")
print(conf_matrix)
print("\nAccuracy:", accuracy)
print("Precision:", precision)
print("Recall:", recall)
print("F1 Score:", f1)
```

The output of K-Means clustering was inconclusive at best. It could not accurately group the dataset as diabetes_mellitus positive and negative.

```python
# Clusters 0 and 1 may or may not necessarily represent diabetes_mellitus as false and true.
# Hence, it is necessary to switch the cluster labels to check the alternate possibility
# Replace 1 with 0 and 0 with 1 in the 'Pred_Cluster' column

# Create a new DataFrame df_reverse by copying df_original 
df_reverse = df_original.copy(deep=True)

# Replace 1 with 0 and 0 with 1 in the 'Pred_Cluster' column of df_reverse
df_reverse['Pred_Cluster'] = df_reverse['Pred_Cluster'].replace({1: 0, 0: 1})

display(df_reverse.head(5))
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score

actual_values = df_reverse['diabetes_mellitus']
predicted_values = df_reverse['Pred_Cluster']  

conf_matrix = confusion_matrix(actual_values, predicted_values)
accuracy = accuracy_score(actual_values, predicted_values)
precision = precision_score(actual_values, predicted_values)
recall = recall_score(actual_values, predicted_values)
f1 = f1_score(actual_values, predicted_values)

print("Confusion Matrix:")
print(conf_matrix)
print("\nAccuracy:", accuracy)
print("Precision:", precision)
print("Recall:", recall)
print("F1 Score:", f1)
```

*Based on the above analysis, we can conclude that K-means clustering is not a reliable technique for identifying diabetes_mellitus positive and negative labels on this dataset, compared to other algorithms we have previously tested.*
