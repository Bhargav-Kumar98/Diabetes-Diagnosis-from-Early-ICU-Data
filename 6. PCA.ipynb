# Importing Data

## Loading and Displaying DataFrame



Importing the necessary libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import display, HTML
import warnings
warnings.filterwarnings("ignore")
```

Reading the CSV file into a dataFrame and displaying the first few rows of the DataFrame to examine its structure and contents

```python
df = pd.read_csv('imputed_dataset.csv')
pd.set_option('display.max_columns', None)
df.head()
```

Checking the dimensions of the dataframe

```python
df.shape
```

Re-evaluating for missing values

```python
column_missing_info = []
for col in df.columns:
  if(df[col].isnull().any()):
    column_missing_info.append({
        'Column': col,
        'Data Type': df[col].dtype,
        'Null values count': df[col].isnull().sum(),
        '% of missing data': ((df[col].isnull().sum())/len(df))*100
    })

if column_missing_info:
  column_info_df = pd.DataFrame(column_missing_info)
  display(HTML(column_info_df.sort_values(by=["Data Type", "Null values count"]).to_html()))
else:
  print('No missing values!')
```

# Data Transformation

Dropping 'hospital_id', 'icu_id' columns as they didn't have any considerable correlation with the target variable

```python
df = df.drop(columns=['hospital_id', 'icu_id'])
df.head()
```

Separating and Storing numerical and categorical columns

```python
numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns
categorical_cols = df.select_dtypes(include=['object']).columns
```

Taking copy of orginal dataframe before making any transformations

```python
df_original = df.copy(deep=True)
```

## Categorical encoding

Performing one-hot encoding on categorical variables and assigning the new encoded columns to the dataframe

```python
discret_df = pd.get_dummies(df[categorical_cols], drop_first = True)
df = pd.concat([df, discret_df], axis=1)
df.drop(categorical_cols, axis=1, inplace=True)
df.replace({True: 1, False: 0}, inplace=True)

display(HTML(df.head().to_html()))
```

Separating numeric columns from binary columns, so that standardization can be applied to numeric columns only

```python
binary_columns = [x for x in df.columns if (df[x].nunique() == 2 and set(df[x].unique()) == {0, 1})]
numeric_cols = [x for x in df.columns if x not in binary_columns]
```

## Performing Data Normalization

Performing Data Normalization to ensure that all numerical features in the dataset have a similar scale

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
df[numeric_cols] = pd.DataFrame(scaler.fit_transform(df[numeric_cols]), columns=numeric_cols)
display(HTML(df.head().to_html()))
```

# Model Training and Evaluation

Separating the target feature from the independent features for model training

```python
y_feature = df['diabetes_mellitus']
x_features = df.drop('diabetes_mellitus',axis=1)
```

## Model - PCA

```python
from sklearn.model_selection import train_test_split
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score
from imblearn.over_sampling import SMOTE
from sklearn.decomposition import PCA

# Iterating over PCA values with a gap of 10
for n_components in range(7, 97, 10):
    print(f"Applying PCA with {n_components} components")
    pca = PCA(n_components=n_components)
    x_features_After_pca = pca.fit_transform(x_features)

    # Splitting the data into train and test sets
    x_train, x_test, y_train, y_test = train_test_split(x_features_After_pca, y_feature, test_size=0.2, random_state=42)

    # Balancing dataset using SMOTE to resolve imbalance Issue
    sm = SMOTE(random_state=42)
    x_resampled, y_resampled = sm.fit_resample(x_train, y_train)

    # Training the XGBoost Classifier
    model = XGBClassifier(
        n_estimators=1500,
        max_depth=3,
        learning_rate=0.1,
        gamma=0,
        min_child_weight=1,
        subsample=0.8,
        colsample_bytree=0.8,
        reg_alpha=0,
        reg_lambda=1
    )
    model.fit(x_resampled, y_resampled)

    # Evaluating the model on the test set
    y_pred = model.predict(x_test)
    model_accuracy = accuracy_score(y_test, y_pred)
    model_precision = precision_score(y_test, y_pred, average='macro')
    model_recall = recall_score(y_test, y_pred, average='macro')
    model_f1 = f1_score(y_test, y_pred, average='macro')
    model_auc = roc_auc_score(y_test, y_pred)

    print('Model performance (Hold-out Validation)')
    print("- Accuracy: {:.4f}".format(model_accuracy))
    print("- Precision: {:.4f}".format(model_precision))
    print("- Recall: {:.4f}".format(model_recall))
    print("- F1-Score: {:.4f}".format(model_f1))
    print("- AUC: {:.4f}".format(model_auc))
    print('='*35)
    print('\n')
```

Comparatively, PCA with 47 components is the best PCA model.
It is noticed that despite increasing the PCA components beyond 47, the increase in model performance is negligible as we are using more PCA's for little gain.

## XGBoost with PCA - Hold Out Evaluation

```python
from sklearn.model_selection import train_test_split
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score
from imblearn.over_sampling import SMOTE
from sklearn.decomposition import PCA


print(f"Applying PCA with 47 components")
pca = PCA(n_components=47)
x_features_After_pca = pca.fit_transform(x_features)

# Splitting the data into train and test sets
x_train, x_test, y_train, y_test = train_test_split(x_features_After_pca, y_feature, test_size=0.2, random_state=42)

# Balancing dataset using SMOTE to resolve imbalance Issue
sm = SMOTE(random_state=42)
x_resampled, y_resampled = sm.fit_resample(x_train, y_train)

# Training the XGBoost Classifier
model = XGBClassifier(
    n_estimators=2500,
    max_depth=3,
    learning_rate=0.1,
    gamma=0.75,
    min_child_weight=1,
    subsample=0.8,
    colsample_bytree=0.9,
    reg_alpha=0,
    reg_lambda=1
)
model.fit(x_resampled, y_resampled)

# Evaluating the model on the test set
y_pred = model.predict(x_test)
model_accuracy = accuracy_score(y_test, y_pred)
model_precision = precision_score(y_test, y_pred, average='macro')
model_recall = recall_score(y_test, y_pred, average='macro')
model_f1 = f1_score(y_test, y_pred, average='macro')
model_auc = roc_auc_score(y_test, y_pred)

print("XGBoost Classifier by using best PCA components")
print('Model performance (Hold-out Validation)')
print("- Accuracy: {:.4f}".format(model_accuracy))
print("- Precision: {:.4f}".format(model_precision))
print("- Recall: {:.4f}".format(model_recall))
print("- F1-Score: {:.4f}".format(model_f1))
print("- AUC: {:.4f}".format(model_auc))
print('='*35)
print('\n')
```

The above XGBoost (Hold-out) model has an accuracy of 75.63%
The F1 score which is a function of precision and recall metrics, has a value of 67.92%
The AUC value which determines the ability of the model to distingusish between positive and negative labels is 70.41%
Hence, the XGBoost (Hold-out) model has reasonably good prediction accuracy.

## XGBoost with PCA - N fold Cross validation

Perform N fold Cross validation on 47 PCA components obtained

```python
from sklearn.model_selection import cross_val_predict, KFold
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score
from imblearn.over_sampling import SMOTE

# Create a 5-fold cross-validation object
kfold = KFold(n_splits=5, shuffle=True, random_state=42)

data_5folds = []

accuracy_list = []
precision_list = []
recall_list = []
f1_list = []
auc_list = []

pca = PCA(n_components=47)
x_features_After_pca = pca.fit_transform(x_features)

for train_index, test_index in kfold.split(x_features_After_pca,y_feature):
    # get actual data by index
    x_train, x_test = x_features.iloc[train_index], x_features.iloc[test_index]
    y_train, y_test = y_feature[train_index], y_feature[test_index]
    # save data into fold
    fold = [x_train, x_test, y_train, y_test]
    # add each fold data into 5folds
    data_5folds.append(fold)

for x_train, x_test, y_train, y_test in data_5folds:
    sm = SMOTE(random_state=42)
    sm.fit(x_train, y_train)
    x_resampled, y_resampled = sm.fit_resample(x_train, y_train)
    model = XGBClassifier(
        n_estimators=1500,
        max_depth=3,
        learning_rate=0.1,
        gamma=0,
        min_child_weight=1,
        subsample=0.8,
        colsample_bytree=0.8,
        reg_alpha=0,
        reg_lambda=1
        )
    model.fit(x_resampled, y_resampled)
    y_pred = model.predict(x_test)
    # Calculate the performance metrics
    model_accuracy = accuracy_score(y_test, y_pred)
    model_precision = precision_score(y_test, y_pred, average='macro')
    model_recall = recall_score(y_test, y_pred, average='macro')
    model_f1 = f1_score(y_test, y_pred, average='macro')
    model_auc = roc_auc_score(y_test, y_pred)
    
    accuracy_list.append(model_accuracy)
    precision_list.append(model_precision)
    recall_list.append(model_recall)
    f1_list.append(model_f1)
    auc_list.append(model_auc)

print('XGBoost Classifier Model performance(5-Fold CV)')
print("- Accuracy: {:.4f}".format(np.mean(accuracy_list)))
print("- Precision: {:.4f}".format(np.mean(precision_list)))
print("- Recall: {:.4f}".format(np.mean(recall_list)))
print("- F1-Score: {:.4f}".format(np.mean(f1_list)))
print("- AUC: {:.4f}".format(np.mean(auc_list)))
print('='*35)
print('\n')
```

The accuracy (83.10%) of the above XGBoost Classifier (5-Fold CV) is the best among all other models. Particularily, the ability of the model to identify the diabetes_mellitus positive labels is high.

However, the lack of model performance with regards to the AUC value (70.56%) is evident. It indicates the shortcomings of the model to distinguish between positive and negative labels.

Overall, The above model is amongst our best performing models
